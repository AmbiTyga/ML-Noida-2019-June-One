```python
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
text = open("../datasets/speeches/speech.txt").read()
```

```python
sents = sent_tokenize(text)
```

```python
words = list(map(word_tokenize, sents))
```
